```python
import cv2
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread("../build/emotions/FER2013Valid/fer0029612.png")

# Show the image
plt.subplot(1, 4, 1)
plt.imshow(image)

# Load an image
image_2 = cv2.imread("../build/emotions/FER2013Valid/fer0031044.png")

# Show the image
plt.subplot(1, 4, 2)
plt.imshow(image_2)

# Load an image
image_3 = cv2.imread("../build/emotions/FER2013Valid/fer0028638.png")

# Show the image
plt.subplot(1, 4, 3)
plt.imshow(image_3)

# Load an image
image_4 = cv2.imread("../build/emotions/FER2013Valid/fer0028645.png")

# Show the image
plt.subplot(1, 4, 4)
plt.imshow(image_4)
```

```python
# Equalize the image to make it clearer
eq_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
eq_image = cv2.equalizeHist(eq_image)
eq_image = cv2.cvtColor(eq_image, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 1)
plt.imshow(eq_image)

# Equalize the image to make it clearer
eq_image_2 = cv2.cvtColor(image_2, cv2.COLOR_BGR2GRAY)
eq_image_2 = cv2.equalizeHist(eq_image_2)
eq_image_2 = cv2.cvtColor(eq_image_2, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 2)
plt.imshow(eq_image_2)

# Equalize the image to make it clearer
eq_image_3 = cv2.cvtColor(image_3, cv2.COLOR_BGR2GRAY)
eq_image_3 = cv2.equalizeHist(eq_image_3)
eq_image_3 = cv2.cvtColor(eq_image_3, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 3)
plt.imshow(eq_image_3)

# Equalize the image to make it clearer
eq_image_4 = cv2.cvtColor(image_4, cv2.COLOR_BGR2GRAY)
eq_image_4 = cv2.equalizeHist(eq_image_4)
eq_image_4 = cv2.cvtColor(eq_image_4, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 4)
plt.imshow(eq_image_4)
```

```python
# Denoise the image to make it easier for model to learn
denoised_image = cv2.cvtColor(eq_image, cv2.COLOR_BGR2GRAY)
denoised_image = cv2.fastNlMeansDenoising(src=denoised_image, h=16, templateWindowSize=7, searchWindowSize=21)
denoised_image = cv2.cvtColor(denoised_image, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 1)
plt.imshow(denoised_image)

# Denoise the image to make it easier for model to learn
denoised_image_2 = cv2.cvtColor(eq_image_2, cv2.COLOR_BGR2GRAY)
denoised_image_2 = cv2.fastNlMeansDenoising(src=denoised_image_2, h=16, templateWindowSize=7, searchWindowSize=21)
denoised_image_2 = cv2.cvtColor(denoised_image_2, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 2)
plt.imshow(denoised_image_2)

# Denoise the image to make it easier for model to learn
denoised_image_3 = cv2.cvtColor(eq_image_3, cv2.COLOR_BGR2GRAY)
denoised_image_3 = cv2.fastNlMeansDenoising(src=denoised_image_3, h=16, templateWindowSize=7, searchWindowSize=21)
denoised_image_3 = cv2.cvtColor(denoised_image_3, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 3)
plt.imshow(denoised_image_3)

# Denoise the image to make it easier for model to learn
denoised_image_4 = cv2.cvtColor(eq_image_4, cv2.COLOR_BGR2GRAY)
denoised_image_4 = cv2.fastNlMeansDenoising(src=denoised_image_4, h=16, templateWindowSize=7, searchWindowSize=21)
denoised_image_4 = cv2.cvtColor(denoised_image_4, cv2.COLOR_GRAY2BGR)

# Show the image
plt.subplot(1, 4, 4)
plt.imshow(denoised_image_4)
```
